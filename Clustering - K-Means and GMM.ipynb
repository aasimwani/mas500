One method often used in clustering tasks is called k-Means. It is an iterative algorithm where we start with a guess for the cluster center-points (means) and continue to refine our guess until the algorithm converges and stops changing. The basic algorithm is pretty simple so we'll implement it directly here.

## 1. choose a K value

Choosing an appropriate K value might be easy if you happen to know ahead of time how many clusters there are in your data. If you don't know ahead of time, there are a number of methods to choose. One common method is called the Bayesian Information Criterion (BIC), which is a general technique in model selection that allows you to assign a score to a model. Conceptually, the BIC score will be higher for models that better fit the data (give the data a higher probability), but the score is penalized for model complexity (measured in the number of free parameters in the model). One result of using the BIC is that in general having more data will afford you a more complex model.

# Gaussian Mixture Models

Gaussian Mixture Models (GMMs) are very similar to K-means clustering. In fact, K-means can be considered a special case of a GMM. The same concept of picking a starting set of parameters and then iterating. In a GMM, each cluster is a gaussian distribution, defined by a mean ($\mu$) and a covariance matrix ($\Sigma$).

## Covariance Matrices

While the idea of the mean of a multivariate gaussian is pretty intuitive, often the covariance is not so immediately clear. In an d-dimensional gaussian distribution, the covariance matrix will be a dxd matrix. Each cell of the matrix is the covariance between the dimensions indexed by the rows and columns. For instance, in 3D, we might have a covariance matrix:

$$
\left[ \begin{matrix} 0.025 & 0.0075 & -0.00175 \\ 0.0075 & 0.0070 & 0.00135 \\ -0.00175 & 0.00135 & 0.00043 \end{matrix}\right]
$$

Which would be a gaussian where the x and y, and y and z dimentions are positively correlated and the x and z dimentions are negatively correlated.

## Thinking Probabilistically

In the K-means we were pretty ad-hoc about why we'd want to move the centers of the clusters, because it seemed to intuitively make sense to do so. In Mixture Model parlance it's helpful to be a little more formal.

Once we've picked initial parameters for our clusters, each one defines a probability distribution. For 3 clusters, let's call them $P_1$, $P_2$, and $P_3$.

What we're doing when we pick which cluster each point belongs to (Step 3 above) is choosing a cluster for each point $x_i \in X$:

$$
\DeclareMathOperator*{\argmax}{argmax}
y_i = \argmax_k \hspace{3 pt} P_k(x_i; \mu_k^{[t]}, \Sigma_k^{[t]})
$$

That is, finding the cluster $i$ for which the point $x$ has the highest probability. Notice that if our distribution is a spherical gaussian, this is equivilant to finding the cluster with the closest mean (sound familiar?). In the literature you'll see this referenced as the "Expectation" step.

The next step is known as the "Maximization" step, where we find the parameters for our model that maximize the probability of our given data. More formally, for each cluster $k$:

$$
\DeclareMathOperator*{\argmax}{argmax}
\begin{align}
\mu_k^{[t+1]}, \Sigma_k^{[t+1]} &= \argmax_{\mu, \Sigma} \hspace{5pt} P_k(X_k; \mu, \Sigma) \\
                                &= \argmax_{\mu, \Sigma} \prod_{x \in X_k} \mathrm{Normal}(x; \mu, \Sigma)
\end{align}
$$

where $X_k$ is the set of $x \in X$ that we've selected in the Expectation step to be in cluster $k$. 

### What assumption is this making?

Notice how the probability of the training data is the product of the training data, and the probability of each training point is Normally distributed. This is only valid if we assume the data points are "i.i.d", which is an acronym that you'll see a lot in the literature that means "Independent and Identically Distributed", which means that the same distribution (Normal with the given $\mu$ an $\Sigma$) applies to all data points, and they are independent of each other.

This is a slightly simplified example of a general algorithm called the "Expectation-Maximization", or "EM" algorithm. It has many variants, but it's always an iterative back-and-forth algorithm. One important variant is that in the E-step, instead of making a hard selection of which cluster each point is in, you actually calculate the expected value of the cluster, which is where the step gets its name.

At each step you can see how we find new values for $\mu_i$ and $\Sigma_i$, for each of our 3 clusters. As with the K-means algorithm it will not necessarily converge to a global optimum, so it's important to run several times with different starting points to more fully explore the space. 

```python
import matplotlib.pylab as pl
import numpy as np
from numpy.linalg import norm
from sklearn import datasets
pylab.rcParams['figure.figsize'] = (7.0, 7.0)

def apply_kmean(X, means):
    new_means = np.zeros(means.shape)
    cluster_count = np.zeros(means.shape[0])
    Y = np.zeros(X.shape[0])
    for i, x in enumerate(X):
        dists = np.array([norm(x - m) for m in means])
        cluster_id = np.argmin(dists)
        Y[i] = cluster_id
        new_means[cluster_id] += x
        cluster_count[cluster_id] += 1
    new_means = np.array([new_means[i] / count for i, count in enumerate(cluster_count)])
    return Y

def recalc_means(X, Y, n_means):
    means = np.zeros([n_means, X.shape[1]])
    cluster_count = np.zeros(n_means)
    for i, x in enumerate(X):
        cluster_id = int(Y[i])
        means[cluster_id] += x
        cluster_count[cluster_id] += 1
    means = np.array([means[i] / count for i, count in enumerate(cluster_count)])
    return means
```
