Pairwise scatter plots are handy tools for visualization high dimensional datasets.

Let's use it to look at the iris dataset:

```python
%matplotlib inline
import matplotlib.pyplot as plt

from sklearn import datasets
```

```python
iris = datasets.load_iris()
```

```python
pair_plot(iris.data)
```

As it says on the can, **pair_plot()** took every pair of columns in the dataset and did a scatter plot of those pairs. The figure is naturally symmetric around the diagonal (with x an y inverted). Plotting a variable against itself will not be very useful so we show a histogram of that variable.

This shows for instance that var3 of iris dataset is highly correleated with var2.

What this plot doesn't show is the class labels (iris.target) for each point. For that we pass in target:

```python
pair_plot(iris.data, iris.target)
```

et voila

```python
%matplotlib inline
import matplotlib.pyplot as plt


def pair_plot(data, target=None, labels=None):
    """ Generate pairwise scatter plot of data with coloring
    by target.
    
    Adapted from stackoverflow.com/questions/2682144 - palash

    """

    nVariables = data.shape[1] # get the number of columns
    if target is not None:
        assert target.shape[0] == data.shape[0]
        
    if labels is None:
        labels = ['var%d'%i for i in range(nVariables)]
    fig = plt.figure(figsize=(12,12))
    for i in range(nVariables):
        for j in range(nVariables):
            nSub = i * nVariables + j + 1
            ax = fig.add_subplot(nVariables, nVariables, nSub)
            if i == j:
                ax.hist(data[:,i], color='#CCCCCC')
                ax.set_title(labels[i])
                ax.spines['right'].set_color('none')
                ax.spines['top'].set_color('none')
                ax.spines['left'].set_color('none')
                
                ax.xaxis.set_ticks_position('bottom')
                ax.yaxis.set_visible(False)
                    
            else:
                if target is None:
                    ax.scatter(data[:,j], data[:,i], c='red', alpha=0.5)
                else:
                    ax.scatter(data[:,j], data[:,i], c=target, alpha=0.5, cmap=plt.cm.hot)
                    
                if i != nVariables - 1:
                    ax.xaxis.set_visible(False)
                if j != 0:
                    ax.yaxis.set_visible(False)
                    
                ax.xaxis.set_ticks_position('bottom')
                ax.yaxis.set_ticks_position('left')
                ax.spines['right'].set_color('#CCCCCC')
                ax.spines['top'].set_color('#CCCCCC')
                ax.spines['left'].set_color('#CCCCCC')
                ax.spines['bottom'].set_color('#CCCCCC')


```
