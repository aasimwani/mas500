# The Perceptron Algorithm

## Classification Revisited

* Focus on binary classification
* Assume we're given some known "training data" with labels (-1 and 1) for each point

## Some Notation

let $X = \{x_1, x_2, \ldots, x_n\}$ be our training data

For these examples $x_i \in R^2$, so it's easier to plot. The math works for higher-dimensional data as well.

$Y = \{y_1, y_2, \ldots, y_n\}$ is the set of labels, so $y_i \in \{-1, 1\}$

The goal is to find a vector $\theta$ so that for some new data point $x$ in class $y$, $\mathrm{sign}(\theta \cdot x) = y$. This can also be expressed as $y (\theta \cdot x) > 0$

The below plot shows random points classified by a decision boundary (dashed black line), which is defined by the vector $\theta$

```python
theta = np.array([-1.0, 2.0])
# generate uniform random points in a 10x10 box
X = (np.random.rand(100, 2) - [0.5, 0.5]) * 10
Y = np.sign(X.dot(theta))

plot_sep(X, Y, theta)
```

# Linear Separability

Data that is linearly separable simply means that there there is a linear hyperplane (or a line for 2D data) that can separate the classes from each other

```python
fig = pl.figure(figsize=(8,4))

np.random.seed(0) # ensure repeatable results

sep1 = (np.random.randn(20, 2) + [-2, -2])
sep2 = (np.random.randn(20, 2) + [2, 2])

nosep1 = (np.random.randn(20, 2) + [-1, -1])
nosep2 = (np.random.randn(20, 2) + [1, 1])

ax1 = fig.add_subplot(1, 2, 1)
ax1.set_title("Linearly Separable")
ax1.scatter(sep1[:, 0], sep1[:, 1], c="red")
ax1.scatter(sep2[:, 0], sep2[:, 1], c="blue")
ax2 = fig.add_subplot(1, 2, 2)
ax2.set_title("Not Linearly Separable")
ax2.scatter(nosep1[:, 0], nosep1[:, 1], c="red")
ax2.scatter(nosep2[:, 0], nosep2[:, 1], c="blue")
```

# The Perceptron Algorithm

Given a set of data and labels, how does one find an appropriate $\theta$ vector? The Perceptron algorithm is one of the earliest algorithms to come up with one, assuming that the data is in fact linearly separable.

```python
N = 400
X = np.r_[(np.random.randn(N/2, 2) + [-2, -2]), (np.random.randn(N/2, 2) + [2, 2])]
Y = np.array([-1] * (N/2) + [1] * (N/2))

theta = np.array([0, 0])
dirty = True
updates = 0

while dirty and updates < 1000:
    dirty = False
    for i in range(N):
        if Y[i] * theta.dot(X[i]) <= 0:
            updates += 1
            dirty = True
            theta = theta + Y[i] * X[i]

print("%s after %d updates" % ("Aborted" if dirty else "Finished", updates))
plot_sep(X, Y, theta)
```

## Step by Step

```python
N = 100
true_theta = np.array([-1.0, 2.0])
# generate uniform random points in a 10x10 box
np.random.seed(0)
X = (np.random.rand(N, 2) - [0.5, 0.5]) * 10
Y = np.sign(X.dot(true_theta))

theta = np.array([0, 0])
dirty = True

fig = pl.figure(figsize=(16, 16))
j = 0

while j < 16 and dirty:
    dirty = False
    for i in range(N):
        if j < 16 and Y[i] * theta.dot(X[i]) <= 0:
            theta = theta + Y[i] * X[i]
            ax = fig.add_subplot(4, 4, j+1, xlim=[-5,5], ylim=[-5,5])
            plot_sep(X, Y, theta, ax, False)
            ax.scatter(X[i, 0], X[i, 1], s=80, facecolors='none')
            dirty = True
            j += 1
```

# Utility Code

The following code is used to simplify the above code. Run this before evaluating any of the cells above.

```python
import matplotlib.pylab as pl
import numpy as np

def plot_sep(X, Y, theta, ax=None, theta_label=True):
    # get the decision boundary
    a = -theta[0] / theta[1]
    sep_x = np.linspace(-5, 5)
    sep_y = a * sep_x

    # draw it
    if ax is None:
        fig = pl.figure(figsize=(8, 8))
        ax = fig.add_subplot(1, 1, 1, xlim=[-5,5], ylim=[-5,5])
        ax.grid(True)
    ax.scatter(X[:, 0], X[:, 1], c=Y, cmap=pl.cm.Paired)
    ax.plot(sep_x, sep_y, "k--")
    ax.quiver([0], [0], theta[0], theta[1], angles='xy', scale_units='xy', scale=1)
    if theta_label:
        ax.text(theta[0], theta[1], r"$\theta$", fontsize=24)
```
