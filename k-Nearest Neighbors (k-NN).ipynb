# k-Nearest Neighbors (k-NN)

K-Nearest Neighbors is a very simple classifier. The general idea is that if a test point is close to more training points of one class than another, it's likely to be part of that class. At the start you pick a K value. To classify a given input vector (AKA test point), you find the K nearest points, and treat them as votes for which class the test point belongs to. One interesting feature of k-NN is that doesn't have an explicit training step.

### Pros

* Fastest possible training (none!)
* simple classification rule

### Cons

* can easily over-fit
* classification speed slows down with large input data-sets (though there are approaches to minimize this)
* very susceptible to the "Curse of Dimensionality"
* sensitive to scaling mis-match between dimensions
* biased towards more prevalent class

```python
# Utility Code

import numpy as np
import matplotlib.pylab as pl

def plot_boundary(clf, n_classes, xlim=None, ylim=None, ax=None):
    if xlim is None:
        xlim = [-5, 5]
    if ylim is None:
        ylim = [-5, 5]
    if ax is None:
        fig = pl.figure(figsize=(8, 8))
        ax = fig.add_subplot(1, 1, 1, xlim=xlim, ylim=ylim)
        ax.grid(True)
    meshsize = 100
    xx, yy = np.meshgrid(np.linspace(xlim[0], xlim[1], meshsize), np.linspace(ylim[0], ylim[1], meshsize))
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    pl.contour(xx, yy, Z, n_classes, colors="black", linestyles='dashed')

def knn_close():
    X = np.array([[0, 0],
                  [1, 1], [-0.5, 1], [3, -2],
                  [2, 2], [1, -1], [3, -3], [-2, -2], [0, 2]])
    Y = [0, 1, 1, 1, 2, 2, 2, 2, 2]
    scatter(X[:, 0], X[:, 1], c=Y, cmap=plt.cm.spectral)
    ax = pl.gca()
    ax.set_ylim(-5, 5)
    ax.set_xlim(-5, 5)
    
knn_close()
```
